```python
from bs4 import BeautifulSoup as BS
import cloudscraper
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import urllib.request
import time
import json
import re
from datetime import datetime
import requests
import keyboard
scraper = cloudscraper.create_scraper(delay=10,   browser={'custom': 'ScraperBot/1.0',})
url = "https://www.forexfactory.com/calendar?week=sep10.2023"
text = scraper.get(url)
text = text.text
soup = BS(text, "html.parser")
rows = soup.find_all("tr", attrs={"class":"calendar__row"})
datedef = ''
timenow1 = ''
while(True):
    data = {}
    jj = 0
    for row in rows:
        nowpm = datetime.now()
        year1 = int(nowpm.strftime("%Y"))
        try:
            currency = row.find('td', class_='calendar__currency').text.strip()
        except:
            continue
        timenow = row.find('td', class_='calendar__cell calendar__time').text.strip()
        if(len(timenow) > 2):
            timenow1 = timenow
        else:
            timenow = timenow1
        try:
            date = row.find('td', class_='calendar__date').find('span', class_='date').text.strip()
            datedef = date
        except:
            date = datedef
        impact = row.find('td', class_='calendar__impact').find('span')
        f = str(impact)
        colors = ["gra","yel", "ora", "red"]
        for color in colors:
            if color in f:
                if(color == colors[0]):
                    impact = 0
                elif(color == colors[1]):
                    impact = 1
                elif(color == colors[2]):
                    impact = 2
                else:
                    impact = 3
        date_parts = date.split(' ')[1:] 
        date = ' '.join(date_parts)
        if(len(date)>6):
            date = date[:6]
        else:
            None
        am_pm = timenow[-2:]
        if(am_pm == 'am'):
            pm = 0
        else:
            pm = 1
        timenow = timenow[:-2]
        timenow = timenow
        if timenow == "Day":
            timenow = "00:01"
            pm = 0
        if timenow == "All D" or timenow == "Tentati" or 'th' in timenow: #'th' in timenow age mojod bashe true age na false
            timenow = "00:01"                                             #true saat 00:01 bar migardanad
            pm = 0
        title = row.find('td', class_='calendar__event').find('span', class_='calendar__event-title').text.strip()
        actual = row.find('td', class_='calendar__actual').text.strip()
        forecast = row.find('td', class_='calendar__forecast').text.strip()
        previous = row.find('td', class_='calendar__previous').text.strip()
        dt = datetime.strptime(f"{date} {timenow}", "%b %d %H:%M")
        dt = dt.replace(year=year1)
        date = dt.strftime("%Y-%m-%d %H:%M:%S")
#         print(row)
        print("Currency:", currency)
        print("title:", title)
        print("Actual:", actual)
        print("Forecast:", forecast)
        print("Previous:", previous)
        print("Date:", date)
        print("impact:", impact)
        print("pm:", pm)
        print('---------------------------------------------------------------------------------------------------')
        data[jj] = {"currency":currency, "impact":impact, "title":title, "actual":actual, "forcast":forecast, "previous":previous, "timeofevent":date, "pm":pm}
        jj = jj + 1
        time.sleep(2)
```
